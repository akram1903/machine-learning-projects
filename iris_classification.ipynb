```python

# didn't get accurate results
def separability1(data):
  """
  params:
    - data = 2d array first index is the class id, second index is the data
    index inside the class

  formula: separability = ( average distance between class means ) / ( total variance of the feature )
  """
  # Calculate class means for the feature
  num_classes = len(data)
  class_means = np.zeros(num_classes)
  totalPoints = np.array(data)
  totalPoints = totalPoints.flatten()

  for i in range(num_classes):
    class_means[i] = np.mean(data[i])

  # Calculate the variance of the entire feature
  total_variance = np.var(totalPoints)

  # Calculate the average distance between class means
  average_distance = 0
  for i in range(num_classes - 1):
    for j in range(i + 1, num_classes):
      average_distance += np.abs(class_means[i] - class_means[j])
  average_distance /= (num_classes * (num_classes - 1)) / 2

  # Separability metric based on distance between means relative to total variance
  separability = average_distance / total_variance

  return separability

# trying standardized difference
def separability2(data):
  """
  Calculate the standardized difference between the class means for the chosen feature.
  You can achieve this by subtracting the mean of all data points in the feature by each class mean and then dividing by the standard deviation of the entire feature.
  Features with larger standardized differences between class means will likely have better separability.

  params:
    - data = 2d array first index is the class id, second index is the data
    index inside the class.

    Note: data is for a specific feature.

  """
  separability = 0
  # Calculate class means for the feature
  num_classes = len(data)
  class_means = np.zeros(num_classes)
  totalPoints = np.array(data)
  totalPoints = totalPoints.flatten()

  for i in range(num_classes):
    class_means[i] = np.mean(data[i])

  totalMean = np.mean(class_means)
  # Calculate the variance of the entire feature
  total_variance = np.var(totalPoints)

  # calculate standardized difference
  for i in range(num_classes):
    separability += (total_variance/(class_means[i]-totalMean))**2

  return separability
```

#The feature producing heighest separability is **petal length**

###theory:
  yhat = (z)
  z = w . x

  loss function: cross entropy
  ( y , yhat ) = - [ y . log( yhat ) + (1 - y) . log( 1 - yhat ) ]

  gradient = ( yhat - y ) * x

###Steps:
  1- select training set and validation set ( 80 to 20 )
  
  2- choose a batch of the training set to compute gradient with

  3- initializing the parameters
  
  4- do the logistic regression

```python
import random
train = 0.8
batch_ratio = 0.5

third_feature = np.array(prm[2])
allData = third_feature.flatten()

# adding label to the array
allData = np.array([allData,[i//50 for i in range(150)]])
allData = allData.T

sampleIndices = random.sample(range(150),int(150*train))

validationSet = np.zeros((len(allData)-len(sampleIndices),2))
i = 0
for index in set(range(len(allData))) - set(sampleIndices):
    validationSet[i]=allData[index]
    i += 1

# print(f'validation set {validationSet}')

training_set_withLabel = np.zeros((int(150*train),2))
batch_withLabel = np.zeros((int(150*batch_ratio*train),2))
i=0
for index in sampleIndices:
  training_set_withLabel[i] = allData[index]

  if i%int(1/batch_ratio) == 0:
    batch_withLabel[i//int(1/batch_ratio)] = allData[index]
  i+=1

# for sample in training_set_withLabel:
#   if sample[1]==0:
#     color = 'red'
#   elif sample[1]==1:
#     color = 'blue'
#   else:
#     color = 'green'
#   plt.scatter(sample[0],sample[1],c=color)

for sample in batch_withLabel:
  if sample[1]==0:
    color = (0,1,1)
  elif sample[1]==1:
    color = (1,0,1)
  else:
    color = (1,1,0)
  plt.scatter(sample[0],sample[1],c=color)
plt.title("training batch")
plt.show()
```

```python
def computeLoss(x,w,y):
  ones = np.ones(len(x))
  x = np.array([x,ones])
  z = w.T @ x
  yhat = 1/(1+np.exp(-z))
  # yhat = np.where(yhat <= 0.5, 0, 1)

  loss = 0

  for yh in yhat:
    if yh == 0:
      loss -= np.sum((1-y)*np.log10(1-yh) )
    else:
      loss -= np.sum( y*np.log10(yh))

  return loss/len(yhat)

def computeGradient(x,w,y):
  z = w.T @ x
  yhat = 1/(1+np.exp(-z))
  # yhat = np.where(yhat <= 0.5, 0, 1)

  gradient = ( yhat - y ) @ x.T

  #  print(f"gradient = {gradient}")
  return gradient

def gradientDecent(batch_withLabel,training_set_withLabel,learningRate,n_iterations):


  w = np.array([1,1])
  ones = np.ones(len(batch_withLabel))
  batch = np.array([batch_withLabel[:,0],ones])

  y = batch_withLabel[:,1]

  for i in range(n_iterations):


    gradient = computeGradient(batch,w,y)
    # print(gradient)
    w = w - learningRate*gradient

    loss = computeLoss(training_set_withLabel.T[0],w,training_set_withLabel.T[1])
    # print(f'loss = {loss}')
    plt.scatter(i,loss,c='g')
  print(f'w = {w}')
  plt.show()

  return w
```

```python
# running the gradient

learningRate = 0.001

for label in batch_withLabel.T[1]:
    if label > 0:
      label = 1

modelParameters = gradientDecent(batch_withLabel,training_set_withLabel,learningRate,100)
print(batch_withLabel)
print(batch_withLabel.shape)
```

```python
# validating the model

def predict(w,validationSet):
  x = validationSet[:,0]
  ones = np.ones(len(x))
  x = np.array([x,ones])
  z = w.T @ x
  yhat = 1/(1+np.exp(-z))
  yhat = np.where(yhat <= 0.5, 0, 1)

  return yhat

def validate(w,validationSet):
  yhat = predict(w,validationSet)
  # y = validationSet[:,1]

  y = np.where(validationSet[:,1]==0,0,1)
  error = np.where(y == yhat,1,0)
  errorRate = error.sum()/len(yhat)
  return f'{errorRate*100:.2f} %'
```

```python
validate(modelParameters,validationSet)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from IPython.display import display, Markdown, Latex
from sklearn.datasets import make_blobs
from matplotlib.widgets import Slider
import logging
logging.getLogger("tensorflow").setLevel(logging.ERROR)
tf.autograph.set_verbosity(0)
```

```python
samples = allData
# print(samples)
X_train = samples.T[0]
y_train = samples.T[1]

softMax_model = Sequential(
    [
        layers.InputLayer(input_shape=1),
        layers.Dense(25, activation = 'relu'),
        layers.Dense(15,activation = 'relu'),
        layers.Dense(3, activation = 'softmax')
    ]
)

softMax_model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    optimizer=tf.keras.optimizers.Adam(0.001),
)

softMax_model.fit(
    X_train,y_train,
    epochs=100
)
```

```python

output = np.argmax(softMax_model.predict(allData.T[0]),axis=1)
print(output)
# print(X_train)

print(y_train)
```

```python
accuracy = np.average(np.where(output==allData.T[1],1,0))
print(f"accuracy = {accuracy*100:.2f}%")
```
